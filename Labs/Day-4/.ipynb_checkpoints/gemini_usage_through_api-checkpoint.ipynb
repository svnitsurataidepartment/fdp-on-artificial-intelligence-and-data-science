```python
"""Use Gemini API with prompts."""
# If you do not have the `google-genai` package installed, you can install it using:
# pip install -q -U google-genai
from google import genai

my_api_key = "your gemini API"
client = genai.Client(api_key=my_api_key)

response = client.models.generate_content(
    model="gemini-2.5-flash", contents="Explain how AI works in a few words"
)
print(response.text)
"""
Output:
AI learns from data to make predictions or decisions.
"""
```

```python
from google import genai
from google.genai import types

my_api_key = "your gemini key"
client = genai.Client(api_key=my_api_key)

print("Using Gemini API with thinking enabled")
# with thinking enabled (default)
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Explain how AI works in a few words"
    )
print(response.text)
# with thinking disabled
print("Using Gemini API with thinking disabled")
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Explain how AI works in a few words",
    config=types.GenerateContentConfig(
        thinking_config=types.ThinkingConfig(thinking_budget=0) # Disables thinking
    ),
)
print(response.text)
```

```python
prompt = """
Classify the text as one of the following categories.
- large
- small
Text: Rhino
The answer is: large
Text: Mouse
The answer is: small
Text: Snail
The answer is: small
Text: Elephant
"""
print("Using Gemini API with thinking enabled")
# with thinking enabled (default)
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
    )
print(response.text)
```
