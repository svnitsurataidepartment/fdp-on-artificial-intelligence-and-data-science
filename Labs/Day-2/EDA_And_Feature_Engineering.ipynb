```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
# 0 is name of column, rename with Rating Count
```

| Part                                   | Description                                                                                        |
| -------------------------------------- | -------------------------------------------------------------------------------------------------- |
| `final_df.groupby([...])`              | Groups the data by **three columns**: `'Aggregate rating'`, `'Rating color'`, and `'Rating text'`. |
| `.size()`                              | Counts the number of records (rows) in each group.                                                 |
| `.reset_index()`                       | Converts the grouped index into regular columns, turning the result into a new DataFrame.          |
| `.rename(columns={0: 'Rating Count'})` | Renames the default count column (`0`) to a more readable name: **`'Rating Count'`**.              |

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

| Parameter              | Description                                                              |
| ---------------------- | ------------------------------------------------------------------------ |
| `x="Aggregate rating"` | X-axis shows different Zomato rating scores (e.g., 0.0 to 5.0).          |
| `y="Rating Count"`     | Y-axis shows how many restaurants received that rating.                  |
| `hue="Rating color"`   | Adds color to the bars based on the associated `Rating color`.           |
| `data=ratings`         | Data source is the `ratings` DataFrame (created earlier with `groupby`). |
| `palette=[blue red..]`        | Specifies a custom color palette for the rating categories.              |


## Observation:

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Count plot with custom color palette
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
plot = sns.countplot(x="Rating color", data=ratings,
                     palette=['blue','red','orange','yellow','green','green'])
# Add value labels on top of bars
for p in plot.patches:
    height = p.get_height()
    plot.annotate(f'{height}',               # label text
                  (p.get_x() + p.get_width() / 2., height),  # x, y position
                  ha='center', va='bottom',  # text alignment
                  fontsize=12)
plt.title("Count of Rating Colors", fontsize=14)
plt.xlabel("Rating Color")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
