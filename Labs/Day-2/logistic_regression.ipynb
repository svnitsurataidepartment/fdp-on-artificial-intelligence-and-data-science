```python
import numpy as np
```

```python
x_data = np.array([2, 3, 4, 5, 6, 7]) # normalization may be necessary to get accurate results
y_data = np.array([0, 0, 0, 1, 1, 1]) # we created a sample data with a rule that if x < 5, y = 0 else y = 1
```

```python
def find_sigmoid(w0, w1, x):
    """Find sigmoid value using the weights and the value of x."""
    return 1 / (1 + np.exp(-(w0 + w1 * x)))
```

```python
def batch_gradient_descent_for_logistic_regression(x, y, w0, w1, lr, tolerance):
    """Use batch gradient descent to update the parameters in logistic regression."""
    prev_w0, prev_w1 = w0, w1
    max_iterations = 10000
    for i in range(max_iterations):
        total_update_in_w0, total_update_in_w1 = 0, 0
        for j in range(x.shape[0]):
            p_j = find_sigmoid(w0, w1, x[j])
            update_in_w0 = p_j - y[j]
            update_in_w1 = (p_j - y[j]) * x[j]
            total_update_in_w0 += update_in_w0
            total_update_in_w1 += update_in_w1
        w0 -= lr * total_update_in_w0  # collect updates from all samples and update the parameters
        w1 -= lr * total_update_in_w1
        if abs(w0 - prev_w0) <= tolerance and abs(w1 - prev_w1) <= tolerance:
            break
        else:
            prev_w0, prev_w1 = w0, w1
    return w0, w1       
```

```python
lr = 0.001
tolerance = 0.001
w0, w1 = 0, 0
w0, w1 = batch_gradient_descent_for_logistic_regression(x_data, y_data, w0, w1, lr, tolerance)
```
