```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("EDA_assignment_customer analytics.csv")
```

```python
# Create one-hot representations for categorical data
one_hot_encoded_data = pd.get_dummies(df, columns = ['Education', 'Marital_Status'])
```

```python
one_hot_encoded_data
```

```python
one_hot_encoded_data.drop(columns=["ID", "Year_Birth", "Dt_Customer"])
```

```python
one_hot_encoded_data.columns
```

```python
# Drop irrelevant columns
relevant_data_frame = one_hot_encoded_data.drop(columns=["ID", "Year_Birth", "Dt_Customer"])
```

```python
relevant_data_frame.columns
```

```python
# Fill 0 if values are not available
relevant_data_frame = relevant_data_frame.replace(np.nan, 0)
# Extract the columns as the target of linear regressor
amount_of_wines = relevant_data_frame["MntWines"]
amount_of_fruits = relevant_data_frame["MntFruits"]
amount_of_meat_products = relevant_data_frame["MntMeatProducts"]
amount_of_fish_products = relevant_data_frame["MntFishProducts"]
amount_of_sweet_products = relevant_data_frame["MntSweetProducts"]
amount_of_gold_products = relevant_data_frame["MntGoldProds"]
```

```python
# Drop a single amount variable from the data, 6 types
relevant_data_frame_without_wine_amount = relevant_data_frame.drop(columns=["MntWines"])
relevant_data_frame_without_fruit_amount = relevant_data_frame.drop(columns=["MntFruits"])
relevant_data_frame_without_meat_amount = relevant_data_frame.drop(columns=["MntMeatProducts"])
relevant_data_frame_without_fish_amount = relevant_data_frame.drop(columns=["MntFishProducts"])
relevant_data_frame_without_sweet_amount = relevant_data_frame.drop(columns=["MntSweetProducts"])
relevant_data_frame_without_gold_amount = relevant_data_frame.drop(columns=["MntGoldProds"])
# Drop all the target variables from the data
relevant_data_frame_without_all_targets = relevant_data_frame.drop(columns=["MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntGoldProds"])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# the following models are trained on data by removing all amount variables
regressor_for_wines_amount = LinearRegression().fit(relevant_data_frame_without_all_targets, amount_of_wines)
regressor_for_fruit_amount = LinearRegression().fit(relevant_data_frame_without_all_targets, amount_of_fruits)
regressor_for_meat_amount = LinearRegression().fit(relevant_data_frame_without_all_targets, amount_of_meat_products)
regressor_for_fish_amount = LinearRegression().fit(relevant_data_frame_without_all_targets, amount_of_fish_products)
regressor_for_sweet_amount = LinearRegression().fit(relevant_data_frame_without_all_targets, amount_of_sweet_products)
regressor_for_gold_amount = LinearRegression().fit(relevant_data_frame_without_all_targets, amount_of_gold_products)
```

```python
# the following models are trained on data by removing one specific amount variable
regressor_for_wines_amount_only = LinearRegression().fit(relevant_data_frame_without_wine_amount, amount_of_wines)
regressor_for_fruit_amount_only = LinearRegression().fit(relevant_data_frame_without_fruit_amount, amount_of_fruits)
regressor_for_meat_amount_only = LinearRegression().fit(relevant_data_frame_without_meat_amount, amount_of_meat_products)
regressor_for_fish_amount_only = LinearRegression().fit(relevant_data_frame_without_fish_amount, amount_of_fish_products)
regressor_for_sweet_amount_only = LinearRegression().fit(relevant_data_frame_without_sweet_amount, amount_of_sweet_products)
regressor_for_gold_amount_only = LinearRegression().fit(relevant_data_frame_without_gold_amount, amount_of_gold_products)
```

```python
regressor_for_wines_amount.coef_
```

```python
# Pedict all the samples through your models
all_wines_amount_predicted = regressor_for_wines_amount.predict(relevant_data_frame_without_target)
all_fruits_amount_predicted = regressor_for_fruit_amount.predict(relevant_data_frame_without_target)
all_meat_amount_predicted = regressor_for_meat_amount.predict(relevant_data_frame_without_target)
all_fish_amount_predicted = regressor_for_fish_amount.predict(relevant_data_frame_without_target)
all_sweet_amount_predicted = regressor_for_sweet_amount.predict(relevant_data_frame_without_target)
all_gold_amount_predicted = regressor_for_gold_amount.predict(relevant_data_frame_without_target)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
# Evaluate the model by taking the root mean square distance for all points
mean_sq_error_wines_amount = sqrt(mean_squared_error(amount_of_wines, all_wines_amount_predicted))
mean_sq_error_fruits_amount = sqrt(mean_squared_error(amount_of_fruits, all_fruits_amount_predicted))
mean_sq_error_meat_amount = sqrt(mean_squared_error(amount_of_meat_products, all_meat_amount_predicted))
mean_sq_error_fish_amount = sqrt(mean_squared_error(amount_of_fish_products, all_fish_amount_predicted))
mean_sq_error_sweet_amount = sqrt(mean_squared_error(amount_of_sweet_products, all_sweet_amount_predicted))
mean_sq_error_gold_amount = sqrt(mean_squared_error(amount_of_gold_products, all_gold_amount_predicted))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate sample data
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Create a K-Means clustering model with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=0)

# Fit the model to the data
kmeans.fit(X)

# Get the cluster labels for each data point
labels = kmeans.labels_

# Get the coordinates of the cluster centers
centroids = kmeans.cluster_centers_

# Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, linewidths=3, color='black')
plt.title('K-Means Clustering')
plt.show()
```
